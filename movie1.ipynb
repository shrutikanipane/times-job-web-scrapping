```python
import requests
from bs4 import BeautifulSoup
import re
url = 'https://www.themoviedb.org'

header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.51 Safari/537.36'}

source_data = requests.get(url+'/movie', headers = header).text

soup_data=BeautifulSoup(source_data,"lxml")

all_divs=soup_data.find_all("div",class_="card style_1")

movie_lst=[]
 
for item in all_divs:#doubt
    movieurl=item.find("a")["href"]
    commonurl="https://www.themoviedb.org"
    
    url1=commonurl+movieurl
    header1 = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.51 Safari/537.36'}

    source_data1 = requests.get(url1, headers = header1)
    #print(source_data1)
    source_data1 = requests.get(url1, headers = header1).text
    soup_data1=BeautifulSoup(source_data1,"lxml")
    
    #first_div=soup_data1.find("div",class_="header_poster_wrapper false")
    first_div=soup_data1.find("section",class_="header poster")
    name=first_div.find("a").text
    name=name.strip()
    
    releasedate=first_div.find("span",class_="release").text
    releasedate=releasedate.strip()
    releasedate=releasedate.split(" ")[0] 
    
    genres=first_div.find("span",class_="genres").text
    genres=genres.strip()
    
    runtime=first_div.find("span",class_="runtime").text
    runtime=runtime.strip()
    
    div1=first_div.find("div",class_="header_info")
    ol=div1.find("ol",class_="people no_image").text
    ol=ol.strip()
    lst=[]
    lst.append(ol)
    print(lst)
    
    
    #pattern="[a-zA-Z]+[\s]+[a-zA-Z]+[\n]+[Director]+[\n]+[\n]"
    #pattern="[a-zA-Z]+[\s]+[a-zA-Z]+[\n]+[Director]+[,a-zA-Z]+[\n]+"
    #pattern="[a-zA-Z]+[\s]+[a-zA-Z]+[\n]+[Director]+"
    pattern="[a-zA-Z]+[\s]+[a-zA-Z]+[\n]+[Director]+.+[\n]*"
    director=re.findall(pattern,ol)
    print(director)
    director="\n".join(director)
    print(director)
    director=director.strip()
    print(director)
    director=director.split("\n")[0]
    print(director)
    print("====================================")
    
    movie_info={
        "name":name,
        "url":url1,
        "releasedate":releasedate,
        "genres":genres,
        "runtime":runtime,
        "director":director  
    }
    
    movie_lst.append(movie_info)
```
